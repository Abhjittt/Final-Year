```python
import cv2
import os

# Paths
input_folder = r"C:\Users\balaj\Desktop\final_year_project\Panoramic radiography database"
output_folder = r"C:\Users\balaj\Desktop\final_year_project\cropped_images"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

def crop_image(image_path, output_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # Allow the user to draw a bounding box
    bbox = cv2.selectROI("Draw Bounding Box", image, fromCenter=False, showCrosshair=True)
    x, y, w, h = bbox
    
    # Crop the image using the bounding box coordinates
    cropped_image = image[int(y):int(y + h), int(x):int(x + w)]
    
    # Save the cropped image
    cv2.imwrite(output_path, cropped_image)
    cv2.destroyAllWindows()

# Loop through all images in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Add more extensions if needed
        image_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, f"cropped_{filename}")
        
        print(f"Processing {filename}...")
        crop_image(image_path, output_path)

print("Cropping completed and saved to the output folder.")
```
